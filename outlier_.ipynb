# z-scores

```python
ul = df['Delivery Rating'].mean()+3*df['Delivery Rating'].std()
ll = df['Delivery Rating'].mean()-3*df['Delivery Rating'].std()
```

```python
ul
```

```python
ll
```

```python
df.loc[(df['Delivery Rating']>ul) | (df['Delivery Rating']<ll)]
```
